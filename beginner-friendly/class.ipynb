```python
class User:

    def __init__(self, user_id, username):
        self.id = user_id
        self.username = username
        self.followers = 0
        self.following = 0
        print("New user being created.")


    def follow(self, user):
        user.followers += 1
        self.following += 1


user = User("001", "Kaya")
user2 = User("002", "Oğlu")

user.follow(user2)

print(user.followers)
print(user.following)
print(user2.followers)
print(user2.following)
```
